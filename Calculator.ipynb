```python
import tkinter as tk
import math

def button_click(number):
    current = entry.get()
    entry.delete(0, tk.END)
    entry.insert(0, str(current) + str(number))

def button_clear():
    entry.delete(0, tk.END)

def button_equal():
    try:
        expression = entry.get()
        result = eval(expression)
        entry.delete(0, tk.END)
        entry.insert(0, str(result))
    except:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")

def button_sqrt():
    try:
        expression = entry.get()
        result = math.sqrt(eval(expression))
        entry.delete(0, tk.END)
        entry.insert(0, str(result))
    except:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")

def button_sin():
    try:
        expression = entry.get()
        result = math.sin(math.radians(eval(expression)))
        entry.delete(0, tk.END)
        entry.insert(0, str(result))
    except:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")

def button_cos():
    try:
        expression = entry.get()
        result = math.cos(math.radians(eval(expression)))
        entry.delete(0, tk.END)
        entry.insert(0, str(result))
    except:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")

def button_tan():
    try:
        expression = entry.get()
        result = math.tan(math.radians(eval(expression)))
        entry.delete(0, tkabs.END)
        entry.insert(0, str(result))
    except:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")

def button_log():
    try:
        expression = entry.get()
        result = math.log10(eval(expression))
        entry.delete(0, tk.END)
        entry.insert(0, str(result))
    except:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")
        
def button_ln():
    try:
        expression = entry.get()
        result = math.log(eval(expression))
        entry.delete(0, tk.END)
        entry.insert(0, str(result))
    except:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")

def button_power():
    current = entry.get()
    entry.delete(0, tk.END)
    entry.insert(0, str(current) + "")

def button_factorial():
    try:
        expression = entry.get()
        result = math.factorial(eval(expression))
        entry.delete(0, tk.END)
        entry.insert(0, str(result))
    except:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")


root = tk.Tk()
root.title("Scientific Calculator")

entry = tk.Entry(root, width=30, borderwidth=5)
entry.grid(row=0, column=0, columnspan=5, padx=10, pady=10)

buttons = [
    '7', '8', '9', '/',
    '4', '5', '6', '*',
    '1', '2', '3', '-',
    '0', '.', '=', '+',
    'C', '√', 'sin', 'cos',
    'tan', 'log', 'ln','x^y','!'
]

row_val = 1
col_val = 0

for button in buttons:
    if button == 'C':
        tk.Button(root, text=button, padx=20, pady=20, command=button_clear).grid(row=row_val, column=col_val)
    elif button == '=':
        tk.Button(root, text=button, padx=20, pady=20, command=button_equal).grid(row=row_val, column=col_val)
    elif button == '√':
        tk.Button(root, text=button, padx=20, pady=20, command=button_sqrt).grid(row=row_val, column=col_val)
    elif button == 'sin':
        tk.Button(root, text=button, padx=20, pady=20, command=button_sin).grid(row=row_val, column=col_val)
    elif button == 'cos':
        tk.Button(root, text=button, padx=20, pady=20, command=button_cos).grid(row=row_val, column=col_val)
    elif button == 'tan':
        tk.Button(root, text=button, padx=20, pady=20, command=button_tan).grid(row=row_val, column=col_val)
    elif button == 'log':
        tk.Button(root, text=button, padx=20, pady=20, command=button_log).grid(row=row_val, column=col_val)
    elif button == 'ln':
        tk.Button(root, text=button, padx=20, pady=20, command=button_ln).grid(row=row_val, column=col_val)
    elif button == 'x^y':
        tk.Button(root, text=button, padx=20, pady=20, command=button_power).grid(row=row_val, column=col_val)
    elif button == '!':
        tk.Button(root, text=button, padx=20, pady=20, command=button_factorial).grid(row=row_val, column=col_val)
    else:
        tk.Button(root, text=button, padx=20, pady=20, command=lambda num=button: button_click(num)).grid(row=row_val, column=col_val)

    col_val += 1
    if col_val > 3:
        col_val = 0
        row_val += 1

root.mainloop()
```
